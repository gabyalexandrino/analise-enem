```python
import chardet
import pandas as pd
```

```python
file = 'dados/MICRODADOS_ENEM_2023.csv'

def detect_encoding(file):
    # Definir o número de bytes a serem lidos (exemplo: 10 MB)
    sample_size = 10 * 1024 * 1024  # 10 MB

    with open(file, 'rb') as f:
        sample = f.read(sample_size)  # Lê apenas os primeiros 10 MB
        result = chardet.detect(sample)
        
        return result['encoding']

print(detect_encoding(file))
```

```python
file_encoding = detect_encoding(file)
df = pd.read_csv('dados/MICRODADOS_ENEM_2023.csv', encoding=file_encoding, sep = ';')
df
```
